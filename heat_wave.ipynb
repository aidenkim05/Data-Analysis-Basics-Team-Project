```python
import pandas as pd
import csv
# 파일 이름들 (직접 업로드한 파일명과 일치해야 함)
file_names = [
    'ISSUE_HW_DAY_2024-05_2024-05_2024.csv',
    'ISSUE_HW_DAY_2024-06_2024-06_2024.csv',
    'ISSUE_HW_DAY_2024-07_2024-07_2024.csv',
    'ISSUE_HW_DAY_2024-08_2024-08_2024.csv',
    'ISSUE_HW_DAY_2024-09_2024-09_2024.csv',
]

# 파일 읽고 병합
df_list = [pd.read_csv(fname, sep='\t', encoding='euc-kr') for fname in file_names]
df = pd.concat(df_list, ignore_index=True)

# 대구 지역만 필터링
df_daegu = df[df['지점'] == '대구(143)']

# 날짜별 최고기온과 폭염 여부 요약 (폭염 기준: 일최고기온 ≥ 33도)
df_daegu['최고기온(°C)'] = pd.to_numeric(df_daegu['최고기온(°C)'], errors='coerce')
df_daegu['폭염여부'] = df_daegu['최고기온(°C)'] >= 33

# 폭염 일수 계산
폭염일수 = df_daegu['폭염여부'].sum()
전체일수 = len(df_daegu)

# 일자별 출력
print("📋 대구 지역 폭염 일별 요약:\n")
for idx, row in df_daegu.iterrows():
    print(f"{row['일시']} - 최고기온(°C): {row['최고기온(°C)']}℃ | 폭염: {'Yes' if row['폭염여부'] else 'No'}")

# 요약 통계
print("\n✅ 전체 분석 결과 요약:")
print(f"- 분석 기간 총 일수: {전체일수}일")
print(f"- 폭염 발생 일수: {폭염일수}일")
print(f"- 폭염 비율: {폭염일수 / 전체일수 * 100:.2f}%")
```
